```python
import pandas as pd
import re
df5 = pd.read_csv("Hotel_Reviews_cleaned_nouns_and_adjs.csv")
print(df5.head())
list_of_nouns=[]
for i in range(df5.shape[0]):
    list_of_nouns.extend(
        re.sub(r'[^a-zA-Z0-9\s]', '', df5['noun_list'][i]).split())
```

```python
list_of_nouns_wdp=list((set(list_of_nouns)))
#sorting list of nouns_wdp in alphabetical order
list_of_nouns_wdp.sort()
print(len(list_of_nouns_wdp))
print(list_of_nouns_wdp[0:50])
final_list = []
for i in list_of_nouns_wdp:
    if not(bool(re.search(r'\d', i)) or len(i)<3 or i.count(i[0]) == len(i)):
        final_list.append(i)
print(len(final_list))
print(final_list[0:50])
```

```python
#lemmatise the list of nouns
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmatized_list = []
for i in Aspects:
    lemmatized_list.append(lemmatizer.lemmatize(i))
print(len(lemmatized_list))
print(len(Aspects))
```

```python
#Bert topic modelling on the list of nouns Aspects
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('bert-base-nli-mean-tokens')
embeddings = model.encode(Aspects)
print(len(embeddings))
print(embeddings[0])
```

```python
#Clustering the aspects using KMeans
from sklearn.cluster import KMeans
import numpy as np
num_clusters = 10
clustering_model = KMeans(n_clusters=num_clusters)
clustering_model.fit(embeddings)
cluster_assignment = clustering_model.labels_
print(cluster_assignment)
```

```python
clustered_sentences = {i: [] for i in range(num_clusters)}
for sentence_id, cluster_id in enumerate(cluster_assignment):
    clustered_sentences[cluster_id].append(Aspects[sentence_id])

for i, cluster in clustered_sentences.items():
    print("Cluster ", i+1)
    #Write each cluster to a pickle file with filename as the cluster number
    with open('cluster'+str(i+1)+'.pickle', 'wb') as handle:
        pickle.dump(cluster, handle, protocol=pickle.HIGHEST_PROTOCOL)
    print(cluster)
    print("")

```

```python
from sklearn.metrics import pairwise_distances
pairwise_dist = pairwise_distances(embeddings)

#Evaluation of the Kmeans clustering
from sklearn.metrics import silhouette_score
silhouette_score(pairwise_dist, cluster_assignment)
print("Silhouette score: ", silhouette_score(embeddings, cluster_assignment))

#Evaluation of the Kmeans clustering
from sklearn.metrics import calinski_harabasz_score
calinski_harabasz_score(embeddings, cluster_assignment)
print("Calinski Harabasz score: ", calinski_harabasz_score(embeddings, cluster_assignment))

#Evaluation of the Kmeans clustering
from sklearn.metrics import davies_bouldin_score
davies_bouldin_score(embeddings, cluster_assignment)
print("Davies Bouldin score: ", davies_bouldin_score(embeddings, cluster_assignment))


```

```python
import spacy
import pandas as pd 
import numpy as np
nlp = spacy.load('en_core_web_lg')
#words = nlp('breakfast food deserts fruits room hashbrowns vegetarians hair bathroom')
words = nlp(words_str)
df= pd.DataFrame()
#cols = ['breakfast','food','deserts','fruits','room','hashbrowns','vegetarians','hair','bathroom']
cols = Aspects
sim = dict.fromkeys(words)
arr = np.empty((len(cols),len(cols)))
print(arr.shape)
arr.fill(1)
k = 0
for i in range(len(cols)-1):
    for j in range(i+1,len(cols)):
        token1 = words[i]
        token2 = words[j]
        temp = token1.similarity(token2)
        arr[i][j] = temp
        arr[j][i] = temp
print(arr)
df2 = pd.DataFrame(arr,columns=cols, index = cols)
print(df2)
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from scipy.spatial.distance import squareform
import matplotlib.pyplot as plt
plt.figure(figsize=(12,5))
dissimilarity = 1 - abs(df2)
Z = linkage(squareform(dissimilarity), 'complete')#

dendrogram(Z, labels=cols, orientation='top',leaf_rotation=90);
```

```python
threshold = 0.5
Numclusters = 500
while threshold <1:
    labels = fcluster(Z, threshold, criterion='distance')
    if(len(np.unique(labels))<Numclusters):
        break
    threshold+=0.001
print(threshold)
# Show the cluster
labels
```
